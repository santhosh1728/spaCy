**Stemming words with NLTK****

stemming is the process of producing morphological variants of a root/base word

changing ->change

changed->chang

change->chang

studing ->studi

studies->studi

study->studi

```python
!pip install nltk
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
ps = PorterStemmer()
```

```python
words = ['changing','changes','changed','change','study','studing']
```

```python
for word in words:
  print(word, ":" , ps.stem(word))
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
for word in words:
  doc=nlp(word)
  for token in doc:
    print(word,"->:",token.lemma_)
```

```python
for word in words:
  doc=nlp(word)
  for token in doc:
    print(word,":",token.lemma_,":",ps.stem(word))
```

**Relu Based Matching in Spacy**
Unlike regular expression's fixed pattern mathing this helps us match token,phase and entities of words and sentences according to some pre-set patterns along with the features such as parts-of-speech,entity types,dependency parsing,lemmatization and may more

There are three kinds of matching methods available as follows


1.   Token matcher
2.   Pharse Matcher
3.    Entity Ruler



**Text Matcher**
it is a token based matching given as matcher,operating over tokens.it makes use of the word level features of spacy such as LOWER,LENGTH,LEMMA,SHAPE and flags such as IS_PUNCT , IS_DIGIT,LIKE_URL,etc.

```python
import spacy
spacy.__version__
```

```python
from spacy.matcher import Matcher
from spacy.tokens import Span # spab helps us to find out the particular piece of the match word in the sentence
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
matcher = Matcher(nlp.vocab)  # Matcher -> is a object ,now we want to merge
```

```python
pattern_verb = [{"LOWER":'alice',"POS":"PROPN"},{"POS":"VERB"}]
pattern = [{"LOWER":'alice',"POS":"PROPN"},{"POS":{"NOT_IN":['AUX']}}]#USED AS THE PROPER NOUN AND FOLLOWED BY PARTS OF HIS SPEECH
```

```python
matcher.add("matching",[pattern_verb])
matcher.add("Matching",[pattern])
```

```python
matches = matcher(doc)
```

```python
for match_id,start,end in matches:
  string_id = nlp.vocab.strings[match_id]
  span = doc[start:end]
  print(match_id,string_id,start,end,span.text)
```
